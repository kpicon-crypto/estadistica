```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
df = pd.read_csv("ResultadosSabanetaSaber11.csv")

# Información básica
print("Información del dataset")
print(df.info())
print("Primeras filas del dataset")
print(df.head())
print("Resumen estadístico del dataset")
print(df.describe())
```

```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())

# Visualización de nulos
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

```python
# Variables categóricas
categorical_cols = ['Año', 'Sector', 'Código DANE', 'Indicador']

for col in categorical_cols:
    print(f"Visualizaciones para la variable: {col}")

    # Diagrama de barras
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col)
    plt.title(f'Distribución de {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Diagrama de torta (solo si el número de categorías no es muy grande)
    if df[col].nunique() < 20: # Limitar para evitar gráficos de torta ilegibles
        plt.figure(figsize=(8, 8))
        df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
        plt.title(f'Proporción de {col}')
        plt.ylabel('') # Eliminar el nombre del eje y para un mejor aspecto
        plt.show()
    else:
        print(f"No se generó diagrama de torta para {col} debido a la gran cantidad de categorías.")
```

```python
# Visualización de la variable numérica 'Resultado'

# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Resultado', kde=True)
plt.title('Distribución de los Resultados')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.show()

# Diagrama de caja (Boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Resultado')
plt.title('Diagrama de Caja de los Resultados')
plt.ylabel('Resultado')
plt.show()
```

```python
# Calcular el primer cuartil (Q1) y el tercer cuartil (Q3)
Q1 = df['Resultado'].quantile(0.25)
Q3 = df['Resultado'].quantile(0.75)

# Calcular el Rango Intercuartílico (IQR)
IQR = Q3 - Q1

# Definir los límites para identificar datos atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar datos atípicos
outliers = df[(df['Resultado'] < lower_bound) | (df['Resultado'] > upper_bound)]

print("Datos atípicos encontrados:")
print(outliers)

# Crear un nuevo DataFrame sin datos atípicos
df_no_outliers = df[(df['Resultado'] >= lower_bound) & (df['Resultado'] <= upper_bound)].copy()

print("\nDimensiones del DataFrame original:", df.shape)
print("Dimensiones del DataFrame sin datos atípicos:", df_no_outliers.shape)
```

```python
from scipy.stats import shapiro, kstest, anderson, jarque_bera
import statsmodels.api as sm

# Asumiendo que 'Resultado' es la variable numérica a evaluar
numerical_var = df['Resultado']

# Shapiro-Wilk Test
shapiro_test = shapiro(numerical_var)
print(f"Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")

# Kolmogorov-Smirnov Test (using a standard normal distribution for comparison)
# Note: For comparing against a specific distribution, you might need to specify parameters
# Here, we compare against a normal distribution fitted to the data's mean and std dev
ks_test = kstest(numerical_var, 'norm', args=(numerical_var.mean(), numerical_var.std()))
print(f"Kolmogorov-Smirnov Test: Statistic={ks_test.statistic:.4f}, p-value={ks_test.pvalue:.4f}")

# Anderson-Darling Test
ad_test = anderson(numerical_var)
print(f"Anderson-Darling Test: Statistic={ad_test.statistic:.4f}")
print("Critical values and significance levels:")
for i in range(len(ad_test.critical_values)):
    print(f"  {ad_test.critical_values[i]:.4f} at {ad_test.significance_level[i]:.2f}%")

# Jarque-Bera Test
jb_test = jarque_bera(numerical_var)
print(f"Jarque-Bera Test: Statistic={jb_test.statistic:.4f}, p-value={jb_test.pvalue:.4f}")

# QQ Plot
sm.qqplot(numerical_var, line='s')
plt.title('QQ Plot for Resultado')
plt.show()
```
